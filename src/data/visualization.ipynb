```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 120

df = pd.read_csv('../../data/raw_data.csv')
print(df.shape)
df.head()
```

```python
# 1. Phân phối final_grade
grade_order = sorted(df['final_grade'].dropna().unique())

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Countplot
sns.countplot(data=df, x='final_grade', order=grade_order, ax=axes[0], palette='Blues_d')
axes[0].set_title('Số lượng học sinh theo final_grade')
axes[0].set_xlabel('final_grade')
axes[0].set_ylabel('Số học sinh')
for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height())}',
                     (p.get_x() + p.get_width() / 2, p.get_height()),
                     ha='center', va='bottom', fontsize=10)

# Pie chart
grade_counts = df['final_grade'].value_counts().reindex(grade_order)
axes[1].pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%',
            colors=sns.color_palette('Blues_d', len(grade_counts)))
axes[1].set_title('Tỉ lệ phần trăm final_grade')

plt.tight_layout()
plt.show()
```

```python
# 2. Phân phối điểm số: math_score, science_score, english_score, overall_score
score_cols = ['math_score', 'science_score', 'english_score', 'overall_score']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, col in enumerate(score_cols):
    sns.histplot(df[col], kde=True, ax=axes[i], color=sns.color_palette('muted')[i], bins=20)
    axes[i].axvline(df[col].mean(), color='red', linestyle='--', linewidth=1.2, label=f'Mean: {df[col].mean():.1f}')
    axes[i].set_title(col)
    axes[i].set_xlabel('Score')
    axes[i].set_ylabel('Count')
    axes[i].legend(fontsize=9)

plt.suptitle('Phân phối điểm số', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
# 3. Boxplot điểm số theo final_grade
grade_order = sorted(df['final_grade'].dropna().unique())

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, col in enumerate(['math_score', 'science_score', 'english_score']):
    sns.boxplot(data=df, x='final_grade', y=col, order=grade_order,
                ax=axes[i], palette='Set2')
    axes[i].set_title(f'{col} theo final_grade')
    axes[i].set_xlabel('final_grade')
    axes[i].set_ylabel(col)

plt.suptitle('Phân phối điểm theo từng grade', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 4. study_hours vs final_grade & overall_score
grade_order = sorted(df['final_grade'].dropna().unique())

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Boxplot study_hours theo final_grade
sns.boxplot(data=df, x='final_grade', y='study_hours', order=grade_order,
            ax=axes[0], palette='Oranges')
axes[0].set_title('study_hours theo final_grade')
axes[0].set_xlabel('final_grade')
axes[0].set_ylabel('study_hours')

# Scatter study_hours vs overall_score, tô màu theo final_grade
sns.scatterplot(data=df, x='study_hours', y='overall_score',
                hue='final_grade', hue_order=grade_order,
                ax=axes[1], palette='tab10', alpha=0.75, s=60)
axes[1].set_title('study_hours vs overall_score')
axes[1].set_xlabel('study_hours')
axes[1].set_ylabel('overall_score')
axes[1].legend(title='final_grade', bbox_to_anchor=(1.01, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# 5. Correlation heatmap
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr = df[num_cols].corr()

fig, ax = plt.subplots(figsize=(10, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, linewidths=0.5, ax=ax, annot_kws={'size': 9})
ax.set_title('Correlation Heatmap (numeric features)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 6. Pairplot: scores + study_hours, màu theo final_grade
pair_cols = ['math_score', 'science_score', 'english_score', 'study_hours', 'final_grade']
pair_df = df[pair_cols].dropna()

g = sns.pairplot(pair_df, hue='final_grade',
                 hue_order=sorted(pair_df['final_grade'].unique()),
                 palette='tab10', plot_kws={'alpha': 0.6, 's': 40},
                 diag_kind='kde')
g.figure.suptitle('Pairplot: Scores & study_hours theo final_grade',
                   y=1.01, fontsize=13, fontweight='bold')
plt.show()
```
